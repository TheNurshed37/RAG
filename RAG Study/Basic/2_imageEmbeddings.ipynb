```python
from transformers import ViTModel, ViTImageProcessor
from PIL import Image
import torch
```

```python
# 1️⃣ Load pre-trained model and processor (feature extractor)
model_name = "google/vit-base-patch16-224-in21k"
processor = ViTImageProcessor.from_pretrained(model_name)
model = ViTModel.from_pretrained(model_name)
```

```python
# 2️⃣ Load and preprocess image
image_path = "1.jpg"  # Change this to your image path
image = Image.open(image_path).convert("RGB")
```

```python
# 3️⃣ Prepare image for model
inputs = processor(images=image, return_tensors="pt")
```

```python
# 4️⃣ Generate embeddings
with torch.no_grad():
    outputs = model(**inputs)
    last_hidden_state = outputs.last_hidden_state      # (batch_size, seq_len, hidden_dim)
    pooled_output = outputs.pooler_output              # (batch_size, hidden_dim)
```

```python
# 5️⃣ Print or use the embeddings
print("Last hidden state shape:", last_hidden_state.shape)
print("Pooled output shape:", pooled_output.shape)
print("Image embedding vector (example):", pooled_output)
```
